```python
# Crear una lista de ligas 
leagues = ['E0','E1','E2','E3','EC']
frames = []

# hacer el loop en la lista league , leer multiples csv y adjuntarlo en un a

for league in leagues:
    df = pd.read_csv( root + '2122' +'/'+ league +'.csv', encoding = 'unicode_escape')
    frames.append(df)
    
 
```

```python
# Crear una lista de ligas 
leagues = ['E0','E2','E3']
frames = []

# hacer loop a un rango de temporadas e insertar una nueva columna 

for league in leagues:
    for season in range(15,21):
        df = pd.read_csv( root + str(season)+  str(season +1)+'/'+ league +'.csv', encoding = 'unicode_escape')
        df.insert(1,'season',season)
        frames.append(df)
```

# Organizar toda la data en un diccionario 

```python
# Crear un diccionario con nombre original de la liga como llave
dict_countries = {'Spanish La Liga':'SP1', 
                  'Spanish Segunda Division':'SP2',
                  'German Bundes':'D1',
                  'English Premier League':'E0','English League 1':'E2','English League 2 ':'E3',
                  }
             
```

```python
dict_countries['Spanish La Liga']
```

```python
dict_historical_data={}
# Crear una lista de ligas 
# leagues = ['E0','E2','E3']


# Hacer loop sobre las llaves del diccionario

for league in dict_countries:
    frames = []
    for season in range(15,21):
        df = pd.read_csv( root + str(season)+  str(season +1)+'/'+ dict_countries[league] +'.csv', encoding = 'unicode_escape')
        df.insert(1,'season',season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
        
        
```

```python
# Mostrar dataframe dentro de dict_historical_data
dict_historical_data['Spanish La Liga']
```

```python
import pandas as pd

# Crear un diccionario para almacenar la información sobre las mayores victorias
dict_mayor_victoria = {'Liga': [], 'Equipo Ganador': []}

# Iterar sobre las ligas en el diccionario
for league in dict_historical_data:
    # Filtrar los partidos de la liga actual
    partidos_liga = dict_historical_data[league]
    
    # Encontrar el partido con la mayor diferencia de goles
    mayor_victoria = partidos_liga.loc[(partidos_liga['FTHG'] - partidos_liga['FTAG']).idxmax()]
    
    # Almacenar la información en el diccionario de mayores victorias
    dict_mayor_victoria['Liga'].append(league)
    dict_mayor_victoria['Equipo Ganador'].append(mayor_victoria['HomeTeam'])

# Crear un DataFrame a partir del diccionario
df_mayor_victoria = pd.DataFrame(dict_mayor_victoria)

# Mostrar el DataFrame de manera más ordenada
for index, row in df_mayor_victoria.iterrows():
    print(f"En la liga {row['Liga']}, el equipo ganador fue: {row['Equipo Ganador']}")
```
